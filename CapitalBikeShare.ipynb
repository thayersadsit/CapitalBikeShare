```python
%matplotlib inline

from sklearn import cross_validation
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime

#Read in Test Files
test=pd.read_csv("test.csv")
train=pd.read_csv("train.csv")
test_date=pd.read_csv("test.csv",parse_dates=["datetime"])
train_date=pd.read_csv("train.csv",parse_dates=["datetime"])


#Create Hours and Month field
train["hour"]=pd.DatetimeIndex(train_date["datetime"]).hour
train["month"]=pd.DatetimeIndex(train_date["datetime"]).month

train.describe()
```

```python
plt.figure()
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(15,8))
train.plot(x='atemp',y='casual',kind='scatter',ax=axes[0,0])
train.plot(x='atemp',y='registered',kind='scatter',ax=axes[0,1])
train.plot(x='temp',y='casual',kind='scatter',ax=axes[1,0])
train.plot(x='temp',y='registered',kind='scatter',ax=axes[1,1])
```

```python
#Determine growth over time on a qualitative basis
plt.figure()
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(15,8))
train.plot(x='month',y='casual',kind='scatter',ax=axes[0,0])
train.plot(x='month',y='registered',kind='scatter',ax=axes[0,1])
train.plot(x='hour',y='casual',kind='scatter',ax=axes[1,0])
train.plot(x='hour',y='registered',kind='scatter',ax=axes[1,1])
```

```python
#initial attempt at random forest before additional features

predictors = ["atemp","hour","workingday","month","holiday","weather","season"]
alg=RandomForestClassifier(random_state=1,n_estimators=30)

alg.fit(train[predictors],train["count"])
train["estimated"]=alg.predict(train[predictors])
train["deviation"]=train["estimated"]-train["count"]
train["deviation2"]=train["deviation"]*train["deviation"]
train["deviation"]=np.sqrt(train["deviation2"])
averageDeviation=sum(train["deviation"])/len(train["deviation"])
print(averageDeviation)
```

```python
#Lets try segregating the riders by casual and registered

predictors = ["atemp","hour","workingday","month","holiday","weather","season"]
alg=RandomForestClassifier(random_state=1,n_estimators=20)

alg.fit(train[predictors],train["casual"])
train["estimatedCasual"]=alg.predict(train[predictors])
alg.fit(train[predictors],train["registered"])
train["estimatedRegistered"]=alg.predict(train[predictors])
train["estimated"]=train["estimatedCasual"]+train["estimatedRegistered"]
train["deviation"]=train["estimated"]-train["count"]
train["deviation2"]=train["deviation"]*train["deviation"]
train["deviation"]=np.sqrt(train["deviation2"])
averageDeviation=sum(train["deviation"])/len(train["deviation"])
print(averageDeviation)
```

```python
#People like to ride bikes if the previous days were cold. Lets take a 3 day average
train["atempMinus1"]=train["atemp"].shift(1)
train["atempMinus1"]=train["atempMinus1"].fillna(train["atemp"])
train["atempMinus2"]=train["atemp"].shift(2)
train["atempMinus2"]=train["atempMinus2"].fillna(train["atemp"])
train["atempMinus3"]=train["atemp"].shift(3)
train["atempMinus3"]=train["atempMinus3"].fillna(train["atemp"])
train["atempMinus3Average"]=(train["atempMinus1"]+train["atempMinus2"]+train["atempMinus3"])/3
train["differenceATemp"]=train["atemp"]-train["atempMinus3Average"]
```
